## A small Introduction 

Many systems are non-linear in nature. Non-linearity makes it hard to understand phenomena. Most assumptions analytically can be sought by assuming linearity or by constraining the system at hand to force linear behaviour. 
The Duffing equation is one of the simplest equations that can exhibit non-linearity. This allows us to qualitatively study many non linear systems - this may be the first step outside of linearization that can catch non linear behaviour of any system. 

Several systems can be modeled by the Duffing Equation. The pendulum is a prime example.  Even more phenomena and systems can be approximated with Duffing characteristics. Isolators, cables, vibrating structures and electric circuits all exhibit Duffing behaviour. The equation also is found in micro and nano mechanical systems, plasmonics, lasers, electric circuits, and so on.

The equation was named after Georg Duffing, a German Engineer, who placed in a cubic non linear term in the vibrations and sought to study them. He wrote a book on the one single equation, almost all of which deals with a standalone Pendulum equation. 



## The Duffing Equation : General notation and various forms 

The most famous version of the equation is 

\begin{equation}
\ddot y + \delta \dot y + \beta y + \alpha y^{3} =  \gamma cos(\omega t)
\end{equation}

where y is the position, derivatives are with respect to time, and $\alpha, \beta, \gamma, \delta$ are all constants. 

We may interpret the equation as an oscillator, or a spring mass system with a non linear stiffness coefficient $$ F = -\beta y - \alpha y^3 $$
This is a Duffing systems which has a harmonic forcing term. When $\alpha < 0$, then the system has a softening spring, and when $\alpha >0$ a hardening spring. Apart from the non-linear stiffness term, all the other terms are identical to forced vibrations in a linear spring-mass system or a pendulum. All terms are in non-dimensional form.

### The Simplest form 

We may add or remove more terms in this equation. The simplest form of the Duffing system would be without the damping term ($\delta = 0$), and the forcing ($\gamma = 0$). 

\begin{equation}
\ddot y + \beta y + \alpha y^{3} =  0
\end{equation}

In many systems, we may neglect damping in favor of the non linear term. There are two terms, a linear stiffness and a non linear stiffness. The takeaway is that stiffness is position dependent.  

### Asymmetry 
The absence of a $y^2$ makes tension and compression in our system symmetric. If the same force does not compress the same amount as it stretches, then we may need to account for an $\it{asymmetry}$ term : 

\begin{equation}
\ddot y + \delta \dot y + \beta y + \epsilon y^{2} + \alpha y^3=  \gamma cos(\omega t)
\end{equation}

## Duffing in a pendulum 

We always make a small anlge approximation to the simple pendulum - this is a form of linearity. However, we may obtain a far more versatile solution (numerically) if we include a Duffing term. 

Consider the equation for a pendulum vertically suspended. The equation for its motion is

\begin{equation}
m\frac{\partial^2 \theta}{\partial t^2} + mgl sin \theta  = M cos \omega t
\end{equation}

Here $M cos \omega t$ is the driving term. If $0$, then we have the natural oscillations of a simple pendulum.  

Now, from the Taylor series, 

$$
sin \theta = \theta - \frac{\theta^3}{3!} + ... 
$$

Substituing into the differential equation, we get, 

\begin{equation}
m\frac{\partial^2 \theta}{\partial t^2} + mgl \left( \theta - \frac{\theta^3}{6} \right )  = M cos \omega t
\end{equation}

And we have a cubic dependent Duffing term in the equation.

# Analysis 

### No damping 

Consider the Duffing Equation in its simplest form - no damping and no forcing, 

\begin{equation}
\ddot y + \beta y + \alpha y^{3} =  0
\end{equation}

We hit each side with $\dot y$. Then, 

\begin{equation}
\dot y \ddot y + \dot y \beta y + \dot y \alpha y^{3} =  0
\end{equation}

Integrating, we get a Conserved Quantity:

\begin{equation}
\frac{1}{2} \dot y ^2 + \frac{\beta}{2} y^{2} + \frac{\alpha}{4}y^{4} = E(t) = const. 
\end{equation}


### Damping 

Now consider the case where there exists some damping $\delta > 0$. On hitting the equation 

\begin{equation}
\ddot y + \delta \dot y + \beta y + \alpha y^{3} =  0
\end{equation}

with $\dot y$, we get :

\begin{equation}
\frac{dE(t)}{dt} = -\delta \dot y^2 \leq 0
\end{equation}

This means that a system with positive damping is always dissipative. The Energy of the system follows up to a state where $\dot y = 0$. The point is an equilibrium point. $(0,0)$ is an equiibrium point of the system. 


## Phase Space analysis 

This will be for the duffing oscillator with no forcing term. Hence, this is an autonomous system with 2 variables, $y$ and $\dot y$. 


We can rewrite the 2nd order Duffing equation as a set of 2 first order ODEs.

\begin{equation}
\ddot y + \delta \dot y + \beta y + \alpha y^{3} =  0
\end{equation}

\begin{equation}
\frac{d}{dt} \begin{bmatrix} y \\ \dot y \end{bmatrix} = \begin{bmatrix} \dot y \\ -\delta \dot y -\beta y -\alpha y^3 \end{bmatrix} = F
\end{equation}

The Jacobian can tell us what the behaviour of the system is very close to a particular point:

\begin{equation}
DF = \begin{bmatrix} 0 & 1 \\ -\beta-3\alpha y^2 & -\delta \end{bmatrix} 
\end{equation}

An equilibrium point is when $F = 0$. That means $\dot y = 0$, and $ -\delta \dot y -\beta y -\alpha y^3 = 0$.
Then:
\begin{equation}
y(\beta + \alpha y^2) = 0
\end{equation}

There are 3 roots for this equation: $y = \pm \sqrt{-\beta/\alpha}$ and $y = 0$. We consider only hardening springs: $\alpha > 0$. For $\beta > 0$, no real roots exist, and the origin is a stable fixed point. This is a single well potential system. 

When $\beta < 0$, the origin becomes an unstable fixed point, and there are 2 stable fixed points. This is then a double well potential system. 

The important thing to note is that there are only 2 variables in phase space. There exists no chaos in this system. This system cannot display long term aperiodic behavior.

# More exciting : The Forced Duffing system 


When there is a periodic forcing term, the system no longer is autonomous - it depends on time, and time becomes the third variable. Hence, we have 3 variables in phase space. 

Consider the equation: 


\begin{equation}
\ddot y + \delta \dot y + \beta y + \alpha y^{3} =  \gamma cos(\omega t)
\end{equation}

The dynamical phase space equations become: 


\begin{equation}
\frac{d}{dt} \begin{bmatrix} y \\ \dot y \\ t \end{bmatrix} = \begin{bmatrix} \dot y \\ -\delta \dot y -\beta y -\alpha y^3 + \gamma cos \omega t\\ 1 \end{bmatrix} = F
\end{equation}

A more convenient way to write down these equations involve using $\theta = \omega t, y_1 = y $ and $y_2 = \dot y$. Using these substitutions, 

\begin{equation}
\begin{bmatrix} \dot y_1 \\ \dot y_2 \\ \dot \theta \end{bmatrix} = \begin{bmatrix} y_2 \\ -\delta y_2 -\beta y_1 -\alpha y_1^3 + \gamma cos \theta \\ \omega \end{bmatrix} = F
\end{equation}

### Equilibrium Points

An equilibrium point happens when $F = \vec 0$ in phase space. However, $F$ is never $\vec 0$, owing to $\dot \theta = \omega = constant$. Hence, there is no equilibrium point in the system. The system can never reach a steady state condition. 



## Solving the ODE using ODEint

```python
# Coefficients 

### ORIGINAL - My own
delta = 0.02
beta = -1
alpha = 5
gamma = 8
omega = 0.5


### PERIODIC 
#delta = 0.2
#alpha = 1
#beta = -1
#gamma = 0.3
#omega = 1

### UEDA 
#delta = 0.05
#gamma = 7.5
#alpha = 1
#beta = 0
#omega = 1

# F
def Duffing(Y, t):
    y1, y2, theta = Y
    return np.array([y2, - delta*y2 - beta*y1 - alpha*y1**3 + gamma*np.cos(theta), omega])

# Initial Conditions  
dt = 1e-3
t = np.arange(0,10000,dt)
Y0 = [1, 0, 0] 

# ODE solve 
from scipy.integrate import odeint 
Duff_sol = odeint(Duffing, Y0, t)

y, y_dot, theta = zip(*Duff_sol)
y = np.array(y); y_dot = np.array(y_dot); theta = np.array(theta) 
```

## Studying Signals of y, $\dot y$, $\theta$

```python
st = int(1000/dt); en = int(1100/dt) # plotting limits
plt.figure(figsize = [12,5]) 
plt.plot(t[st:en],y[st:en]); plt.xlabel('t'); plt.ylabel('y')

plt.figure(figsize = [12,5])
plt.plot(t[st:en],y_dot[st:en]); plt.xlabel('t'); plt.ylabel('$\dot y$')

plt.figure(figsize = [12,5])
plt.plot(t[st:en],theta[st:en]); plt.xlabel('t'); plt.ylabel(r'$\theta$')
```

#### The Signals

* Rate of change of $\theta$ is constant = $\omega$. Hence, $\theta$ is a linear function $\theta = 0.5t$. 


* The $y$ vs $t$ signal shows character. If the signal crosses 0, then, there are multiple oscillations in the $y_+$ region, around 2-3 cycles. Then, the signal breaks away from the region, comes into a transition zone where there is always a kink - a special movement, and then falls into a $y_-$ region where similar oscillations happen. There are 3 main regions where strong activity happens : a $y_+$ region, a $y_-$ region and a transition region where _jumping_ between these 2 regions takes place.


* These regions are akin to well potentials - _energy_ explains what is happening. From the plot, there are 2 potential wells where the signal gets _trapped_ until some condition is satisfied that makes it jump out. 


* Consider a mechanical analogy with spring mass system. There is a damping term that causes the body to lose kinetic energy and velocity. The forcing may not be in phase, hence may not add energy into the system. This causes the system to oscillate small in the $y_+$ and $y_-$ regions.


* However, there may reach a condition when the forcing term starts getting in phase with the velocity, and both achieve a near resonant criterion. Then, the forcing adds in energy, and the body jumps out with an increase in amplitude. Then, damping acts again, the body loses energy and gets trapped in wells and shows small amplitude yet again. 


* Comparing the velocity sginal and the position signal, whenever we enter a transition region, the velocity shoots up, oscillates strongly in one of the wells, and then decreases drastically in amplitude sue to damping, and when force and velocity are in phase, shoots up again - an energy influx.

### 2D phase space $y, \dot y$


* The left and right halves are very similar in character. They are both egg shaped structures. There are 2 eyes, that can be associated with 2 potential wells. 


* Transition between $y_+$ and $y_-$ wells are clearly seen in this figure. The larger orbits of one cell either go round and stay within the same cell, or transit into smaller orbits of another well. There are also obrits which just sneak into another cell, and return back into the same well and then go round again. There is a connection only between small scale trajectories of one well and large scale trajectories of another. 


* There are different rings, or bands of trajectories, and lots of spaces in between. Calling them orbits qualifies to a large extent. Trajectories in each band are very close together for some time. Then, they separate, widen, and then exponentially drift to meet another orbit. Thus, there is clearly a chaotic character present. 


* There is a definite fractal character to the figure. No matter how much we fill it up, we will always find a lot of gaps where trajectories have not passed through yet.


* The cells wind in the same direction. Trajectories and wind up in the same direction throughout. Clockwise winding does not become anticlockwise winding and vice versa, unlike the Lorenz system. 


* The centers of these wells can be seen as chaotic attractors, and they appear to be equilibrium points. However, there are no equilibrium points in the system. Due to external forcing, the trajectories come near the centers, go round and then exit. 


* Large scales mean powerful oscillations with large amplitudes, and small ones mean minute disturbances.

## Poincarè Sections

### First Section 

To obtain some meaning and character to the above plots, we can make Poincarè sections, pertaining to some condition. From the plot $y$ vs $\dot y$, the most meaningful information we can obtain conveniently is when $\dot y = 0$. A plane that cuts through this would give us information as to what is happening with velocity reaches 0. 

The displacement is either a local maximum or a minimum when $\dot y = 0$. We wish to know explicitly what happens when there is a given change in direction of _displacement_, say changing sign from $y^+$ to $y^-$. In other words, concentrate on how maxima behave, or minima behave. 

This section tells us how the long term periodicity of the system turn out, the winding number and the period cycle - almost a kind of frequency spectrum.




```python
# First Poincarè section y. = 0 event locator 
# when y_dot just crosses 0 and acc< 0 : maxima in y

acc = lambda y, ydot, th : -delta*ydot -beta*y -alpha*y**3 + gamma*np.cos(th)

section = []
for x,xp in zip(Duff_sol[:-1],Duff_sol[1:]):
    if (xp[1]>0)!=(x[1]>0) and acc((x[0]+xp[0])/2, (x[1]+xp[1])/2, (x[2]+xp[2])/2) < 0:
        section += [[(x[0]+xp[0])/2, (x[2]+xp[2])/2]]

# plotting        
plt.figure(figsize = [8,8])
section = list(zip(*section))

plt.plot(section[:][0], section[:][1],'o', markersize = 0.5)
plt.xlabel('y'); plt.ylabel(r'$\theta$'); plt.title(r'$\dot y = 0$ and maxima')
```

We see that there exists a periodic solution after a certain time - chaos stops suddenly, and a single periodic orbit emerges

### Second Poincaré section $ \theta = constant$

The most prominent Poincaré section discussed in literature is plotting $y, \dot y$ when the cycle reaches multiples of $2\pi$. We find that there is a fixed point in the system, and all nearby points get spiralled into it. 

```python
# Second Poincare section

def poincare_angle(j):
    th = j*np.pi/3
    section1 = []
    if th == 2*np.pi:
        for x,xp in zip(Duff_sol[:-1],Duff_sol[1:]):
            if x[2]%(2*np.pi) > xp[2]%(2*np.pi): # crossed 
                section1 += [[(x[0]+xp[0])/2, (x[1]+xp[1])/2]]
            
    else:
        for x,xp in zip(Duff_sol[:-1],Duff_sol[1:]):
            if x[2]%(2*np.pi) < th and xp[2]%(2*np.pi) > th: # crossed 
                section1 += [[(x[0]+xp[0])/2, (x[1]+xp[1])/2]]
            
    section1 = list(zip(*section1))
    return section1

fig, ax = plt.subplots(2,3,figsize = [14,10])
fig.suptitle('Poincarè Sections at various angles')
for a in ax.flat:
    a.set(xlabel='y', ylabel= r'$\dot y$')
    a.label_outer()

for j,a in enumerate(ax.flat):
    sec = poincare_angle(j+1)
    a.plot(sec[0][:],sec[1][:],'.',markersize = 3)
```

### Third Poincaré section 

Here, we take a cut-section where $y = 0$. That would give us a sense of what happens when there is a transition from one deep well into another. 

```python
# Third Poincare Section 
# y = 0 and v > 0 

section_y = [] # when y is being eliminated 
for x,xp in zip(Duff_sol[:-1], Duff_sol[1:]):
    if (x[0]>0 and xp[0]<0): # why 
        section_y += [[(x[1]+xp[1])/2, (x[2]+xp[2])/2]]
                                     
section_y = list(zip(*section_y))
plt.figure(figsize = [8,8])
plt.plot(section_y[:][0], section_y[:][1],'o', markersize = 1)
plt.ylabel(r'$\theta$'); plt.xlabel(r'$\dot y$')                      
plt.title('Poincaré section y = 0')                      
                      
```
